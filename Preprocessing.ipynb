```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset.csv', sep=';')
df.head()
```

# Columns Renaming

```python
df.columns
```

```python
rename = {}

oldName = df.columns

newName = [
    'Jurusan',
    'Jenis transportasi yang digunakan',
    'Permasalahan fasilitas parkir',
    'Keamanan fasilitas parkir',
    'Biaya parkir',
    'Kepuasan fasilitas parkir',
    'Tangga/lift yang paling sering dipakai',
    'Alasan menggunakan fasilitas tangga/lift',
    'Kepuasan terhadap fasilitas tangga/lift',
    'Lokasi beli makanan',
    'Alasan memilih beli di tempat itu',
    'Kepuasan fasilitas kantin',
    'Frekuensi penggunaan discussion room',
    'Permasalahan fasilitas discussion room',
    'Kepuasan fasilitas discussion room',
    'Kecepatan dan kestabilan wifi',
    'Kepuasan fasilitas internet',
    'Kapasitas toilet',
    'Kebersihan toilet',
    'Kepuasan fasilitas toilet',
    'Kepuasan secara umum'
]

for i in range(len(oldName)):
    rename[oldName[i]] = newName[i]
```

```python
df.rename(columns=rename, inplace=True)
df.columns
```

# Splitting

```python
splitting = df['Permasalahan fasilitas parkir'].copy().str.split(', ')

df['Permasalahan fasilitas parkir 1'] = splitting.str[0]
df['Permasalahan fasilitas parkir 2'] = splitting.str[1]
```

```python
splitting = df['Permasalahan fasilitas discussion room'].copy().str.split(', ')

df['Permasalahan fasilitas discussion room 1'] = splitting.str[0]
df['Permasalahan fasilitas discussion room 2'] = splitting.str[1]
df['Permasalahan fasilitas discussion room 3'] = splitting.str[2]
```

```python
df.head()
```

```python
df.isnull().sum()
```

Pilihan 'Aksesnya lebih banyak' dan 'Karena bisa loncat-loncat' dihilangkan untuk mengurangi kategori

```python
df['Jenis transportasi yang digunakan'].replace("Buroq", "Motor pribadi", inplace=True)
df['Permasalahan fasilitas parkir 2'].replace("bayar nya mahal", np.nan, inplace=True)

df['Alasan menggunakan fasilitas tangga/lift'].replace(["Aksesnya lebih banyak", "Karena bisa loncat loncat"], [np.nan, np.nan], inplace=True)
```
