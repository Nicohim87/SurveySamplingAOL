```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('dataset_preprocessed.csv', sep=',')
```

```python
df.columns
```

```python
#Kepuasan Parkir
from collections import Counter
counter = Counter(df['Kepuasan fasilitas parkir'])
scores = sorted(counter.keys())
counts = [counter[score] for score in scores]

total_count = sum(counts)
percentages = [(count / total_count) * 100 for count in counts]

labels = [f'Score {score}' for score in scores]

plt.pie(percentages, labels=labels, autopct='%1.1f%%')
plt.title("Kepuasan mahasiswa terhadap fasilitas parkir")
```

```python
#TANGGA LIFT
count = df.groupby(['Tangga/lift yang paling sering dipakai', 'Alasan menggunakan fasilitas tangga/lift']).size().unstack(fill_value=0)

categories = count.index
subcategories = count.columns
bar_width = 0.35
n_subcategories = len(subcategories)

positions = np.arange(len(categories))

fig, ax = plt.subplots(figsize=(12, 8))

for i, subcategory in enumerate(subcategories):
    # Calculate the position for each subcategory's bars
    bar_positions = positions + i * bar_width / n_subcategories
    values = count[subcategory]
    ax.bar(bar_positions, values, bar_width / n_subcategories, label=subcategory)

ax.set_xticks(positions + bar_width / (2 * n_subcategories))
ax.set_xticklabels(categories)

ax.set_title('Jenis fasilitas yang digunakan sebagai sarana menuju kelas')
ax.set_xlabel('Jenis fasilitas')
ax.set_ylabel('Count')
ax.legend(title='Alasan memilih fasilitas')

plt.show()
```
