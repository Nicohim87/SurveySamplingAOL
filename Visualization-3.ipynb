# Violinplot

```python
data = pd.melt(df, id_vars=['Jurusan'], value_vars=[
    'Kepuasan fasilitas discussion room', 'Kepuasan fasilitas internet',
    'Kepuasan fasilitas kantin', 'Kepuasan fasilitas parkir',
    'Kepuasan fasilitas toilet', 'Kepuasan fasilitas tangga/lift',
    'Kepuasan secara umum'
], var_name='Type', value_name='Value')

data
```

```python
sns.violinplot(data, y="Value", hue="Type", width=0.7, dodge=True)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Jenis')
plt.grid(axis='y', linestyle=':')
plt.title("Kepuasan seluruh mahasiswa terhadap semua fasilitas")
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.violinplot(data, y="Value", x="Jurusan", hue="Type", width=0.7, dodge=True)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Jenis')
plt.grid(axis='y', linestyle=':')
plt.title("Kepuasan mahasiswa jurusan tertentu terhadap semua fasilitas")
plt.show()
```

# Corr

```python
sns.heatmap(pd.crosstab(df['Kepuasan fasilitas internet'][df['Jurusan'] == "Computer Science"], df['Kepuasan fasilitas discussion room'][df['Jurusan'] == "Computer Science"]), annot=True)
```

```python
internet = df.groupby('Kepuasan fasilitas internet').size().reset_index(name="total")
merged = df.groupby(['Kepuasan fasilitas internet', 'Kepuasan fasilitas discussion room']).size().reset_index(name="count")

merged = merged.merge(internet, on="Kepuasan fasilitas internet", how="left")
merged['perc'] = merged.apply(lambda row:row['count']/row['total'], axis=1)
merged.head()
```

```python
pivot_df = merged.pivot_table(index='Kepuasan fasilitas internet', columns='Kepuasan fasilitas discussion room', values='perc', fill_value=0)

pivot_df.plot(kind='bar', stacked=True, colormap='viridis', figsize=(10, 7))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Kepuasan fasilitas discussion room')
plt.title("Hubungan kepuasan internet dengan discussion room")
plt.ylabel("Persentase")
plt.yticks([])
plt.plot()
```

```python
mobil = df[df["Jenis transportasi yang digunakan"] == "Mobil pribadi"]


parkir1 = mobil[['Biaya parkir', 'Permasalahan fasilitas parkir 1']].rename(columns={'Permasalahan fasilitas parkir 1':'Permasalahan fasilitas parkir'})
parkir2 = mobil[['Biaya parkir', 'Permasalahan fasilitas parkir 2']].rename(columns={'Permasalahan fasilitas parkir 2':'Permasalahan fasilitas parkir'})


parkir = pd.concat([parkir1, parkir2], ignore_index=True)

parkir.dropna(inplace=True)
parkir
```

```python

mobil = parkir.groupby(['Biaya parkir', 'Permasalahan fasilitas parkir']).size().reset_index(name="count")

mobil['Biaya parkir'].replace(['Iya', 'Tidak'], ['Mahal', 'Tidak'], inplace=True)

pivot_df = mobil.pivot_table(index='Biaya parkir', columns='Permasalahan fasilitas parkir', values='count', fill_value=0)

pivot_df.plot(kind='bar', stacked=True, colormap='viridis', figsize=(10, 7))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Permasalahan fasilitas parkir')
plt.title("Hubungan biaya parkir dengan permasalahannya")
plt.ylabel("Jumlah")
plt.yticks([])
plt.xticks(rotation=0)
plt.plot()
```

```python
stay = df.groupby(['Kecepatan dan kestabilan wifi', 'Lokasi beli makanan', 'Frekuensi penggunaan discussion room']).size().reset_index(name="count")

stay['Frekuensi penggunaan discussion room'].replace(['Ya', 'Tidak'], ['Sering', 'Jarang'], inplace=True)

g = sns.FacetGrid(stay, 
                  hue='Lokasi beli makanan', hue_order=['Di dalam kampus', 'Di luar kampus'],
                  col='Frekuensi penggunaan discussion room', col_order=['Sering', 'Jarang'],
                  height=5, aspect=1)
g.map(sns.barplot, 'Kecepatan dan kestabilan wifi', 'count', order=['Sudah', 'Belum'])
g.add_legend(title="Lokasi beli makanaan")
plt.show()
```
