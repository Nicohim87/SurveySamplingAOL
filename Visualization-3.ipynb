# Violinplot

```python
data = pd.melt(df, id_vars=['Jurusan'], value_vars=[
    'Kepuasan fasilitas discussion room', 'Kepuasan fasilitas internet',
    'Kepuasan fasilitas kantin', 'Kepuasan fasilitas parkir',
    'Kepuasan fasilitas toilet', 'Kepuasan fasilitas tangga/lift',
    'Kepuasan secara umum'
], var_name='Type', value_name='Value')

data
```

```python
sns.violinplot(data, y="Value", hue="Type", width=0.7, dodge=True)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Jenis')
plt.grid(axis='y', linestyle=':')
plt.title("Kepuasan seluruh mahasiswa terhadap semua fasilitas")
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.violinplot(data, y="Value", x="Jurusan", hue="Type", width=0.7, dodge=True)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Jenis')
plt.grid(axis='y', linestyle=':')
plt.title("Kepuasan mahasiswa jurusan tertentu terhadap semua fasilitas")
plt.show()
```

# Corr

```python
sns.heatmap(pd.crosstab(df['Kepuasan fasilitas internet'][df['Jurusan'] == "Computer Science"], df['Kepuasan fasilitas discussion room'][df['Jurusan'] == "Computer Science"]), annot=True)
```

```python
internet = df.groupby('Kepuasan fasilitas internet').size().reset_index(name="total")
merged = df.groupby(['Kepuasan fasilitas internet', 'Kepuasan fasilitas discussion room']).size().reset_index(name="count")

merged = merged.merge(internet, on="Kepuasan fasilitas internet", how="left")
merged['perc'] = merged.apply(lambda row:row['count']/row['total'], axis=1)
merged.head()
```

```python
pivot_df = merged.pivot_table(index='Kepuasan fasilitas internet', columns='Kepuasan fasilitas discussion room', values='perc', fill_value=0)

pivot_df.plot(kind='bar', stacked=True, colormap='viridis', figsize=(10, 7))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Kepuasan fasilitas discussion room')
plt.title("Hubungan kepuasan internet dengan discussion room")
plt.ylabel("Persentase")
plt.yticks([])
plt.plot()
```
